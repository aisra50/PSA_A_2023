```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_ivs = pd.read_excel('dados2.xlsx')
```

```python
df_ivs.columns
```

```python
plt.rcParams["figure.figsize"] = [15.5, 5]
```

```python
x = df_ivs.query("Ano==2010")['Nome da UF']
y = df_ivs.query("Ano==2010").IVS

x1 = df_ivs.query("Ano==2015")['Nome da UF']
y1 = df_ivs.query("Ano==2015").IVS

x2 = df_ivs.query("Ano==2020")['Nome da UF']
y2 = df_ivs.query("Ano==2020").IVS

plt.plot(x, y, 'ro', color = '#DAF7A6', label ='2010')
plt.plot(x1, y1, 'ro', color = '#FFC300', label ='2015')
plt.plot(x2,y2, 'ro', label ='2020' )
plt.xticks(rotation=90)
plt.legend()
plt.title('Índice de Vulnerabilidade Social')
plt.show()
```

```python
x = df_ivs.query("Ano==2010")['Nome da UF']
y = df_ivs.query("Ano==2010").IDHM

x1 = df_ivs.query("Ano==2015")['Nome da UF']
y1 = df_ivs.query("Ano==2015").IDHM

x2 = df_ivs.query("Ano==2020")['Nome da UF']
y2 = df_ivs.query("Ano==2020").IDHM

plt.plot(x, y, 'ro', color = 'pink' , label ='2010')
plt.plot(x1, y1, 'ro', color = '#993DF5', label ='2015')
plt.plot(x2,y2, 'ro', color = '#3D3DF5' , label ='2020' )
plt.xticks(rotation=90)
plt.legend()
plt.title('Índice de Desenvolvimento Humano Municipal por Estado')
plt.show()
```
